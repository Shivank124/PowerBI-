```python
# Top Honey Producing States
df_2 = df.sort_values("production", ascending = False)
```

```python
# let's visualise the above finding
plt.figure(figsize = (30,30))
sns.barplot(x = df_2['production'], y=df_2['state'])
plt.xlabel("Production of Honey")
plt.ylabel("Honey Producing State")
plt.show()
```
